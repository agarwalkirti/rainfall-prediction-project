```python
#import libraries
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('austin_weather_final.csv')
```

```python
data
```

```python
x = data.drop(['PrecipitationSumInches'],axis=1)
```

```python
y = data['PrecipitationSumInches']
```

```python
#reshaping into 2d vector
y=y.values.reshape(-1,1)
#Meaning of parameters in .reshape(-1, 1):-1 → NumPy automatically figures out this dimension based on the data size.
#1 → You explicitly want 1 column.So effectively, you are converting a 1D vector into a column vector.
```

```python
y
```

```python
day_index = 798
days = [i for i in range(y.size)]
```

```python
#initialize the linear regression classifier
model = LinearRegression()
```

```python
# Train Test Split
# y not their in data set so cannot run ML algo on embeddings
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
#train the classifier
model.fit(X_train, y_train)
```

```python
#prediction
y_pred=model.predict(X_test)
```

```python
#Performance metrics and accuracy
y_pred=model.predict(X_test)
```

```python
y_pred #values in terms of probabilities
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

print("MSE:", mean_squared_error(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
#Interpretation:
#MAE is small → good, your model isn’t making very large errors.
#But R² = 0.25 → means most variance is not explained. Basically, your model is only slightly better than predicting the mean rainfall every day.
#confusion matrix,accuracy,classification report used for dicrete classes not continuous predicted values
#from sklearn.metrics import confusion_matrix
#confusion_matrix(y_test,y_pred) #it expects both y_test and y_pred to be discrete class labels, not one continuous and the other binary.
```

```python
#plotting a graph of precipitation levels and total number of days
print('The precipitation trend graph:')
plt.scatter(days,y,color='g',label="Precipitation")
plt.scatter(days[day_index],y[day_index],color='r',label="Highlighted Day")
plt.legend()
plt.title('Precipitation Level')
plt.xlabel('Days')
plt.ylabel('Precipitation in inches')
plt.show()
```
